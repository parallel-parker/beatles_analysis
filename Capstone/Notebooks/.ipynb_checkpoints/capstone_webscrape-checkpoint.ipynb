```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
```

```python
## Because Paul and John always credited their music simply as Lennon/Mccartney, we must
## rely on the research from Beatles superfans who have attempted to sus out which Beatle took
## the lead on each song. We'll use a webpage from an unofficial fansite for this project.
URL = 'http://www.beatlesarchive.net/composer-singer-beatles-songs.html'

response = requests.get(URL)
```

```python
soup = BS(response.text)
```

```python
soup
```

```python
soup.findAll('table')
```

```python
pd.read_html(str(soup.findAll('table')))[0]
```

```python
song_credits = pd.concat(pd.read_html(str(soup.findAll('table'))))
```

```python
song_credits
```

```python
song_credits.drop_duplicates(subset='Song', keep = 'first', inplace = True)
```

```python
song_credits
```

```python
song_credits.replace("´","'")
```

```python
song_credits.to_csv("credits.csv", sep = ',')
```
