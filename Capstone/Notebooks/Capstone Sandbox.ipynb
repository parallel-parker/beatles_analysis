```python
# for loop to get every song name & id from every album

track_ids = []
track_names = []

for i in album_uri:
    tracks = sp.album_tracks(i)
    for t in range(len(tracks['items'])):
        track_ids.append(tracks['items'][t]['id'])
        track_names.append(tracks['items'][t]['name'])        
```

```python
## for loop to get track information for artist

album = []
track_number = []
popularity = []
danceability = []
valence = []

for i in range(len(track_ids)):
    info = sp.track(track_ids[i])
    features = sp.audio_features(track_ids[i])
    album.append(info['album']['name'])
    track_number.append(info['track_number'])
    popularity.append(info['popularity'])
    danceability.append(features[0]['danceability'])
    valence.append(features[0]['valence'])
```
